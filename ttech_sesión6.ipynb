# Seaborn

```python
import pandas as pd

data = pd.read_csv('FuelEfficiency.csv')

data
```

```python
cambios = data['# Gears'].value_counts()

cambios.plot(kind='bar')
```

```python
import seaborn as sns

sns.set()
```

```python
cambios.plot(kind='bar')
```

```python
sns.distplot(data['CombMPG'])
```

```python
consumo_comb = data[['Cylinders','CityMPG','HwyMPG','CombMPG']]
consumo_comb
```

```python
sns.pairplot(consumo_comb)
```

```python
sns.scatterplot(x='Eng Displ',y='CombMPG',data=data)
```

```python
sns.jointplot(x='Eng Displ',y='CombMPG',data=data)
```

```python
sns.lmplot(x='Eng Displ',y='CombMPG',data=data)
```

```python
sns.set(rc={'figure.figsize':(16,5)})

eje_x = sns.boxplot(x='Mfr Name', y='CombMPG', data=data)

eje_x.set_xticklabels(eje_x.get_xticklabels(), rotation=90)
```

```python
eje_x = sns.swarmplot(x='Mfr Name', y='CombMPG', data=data)

eje_x.set_xticklabels(eje_x.get_xticklabels(), rotation=90)
```

# Smart Home Energy Usage

```python
import pandas as pd

data = pd.read_csv('smart_home_energy_usage_dataset.csv')

data
```

```python
data['appliance'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
data.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['energy_consumption_kWh'], bins=10, color='blue', edgecolor='black')
plt.title('DistribuciÃ³n del Consumo de EnergÃ­a (kWh)')
plt.xlabel('Consumo de EnergÃ­a (kWh)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
appliance_energy = data.groupby('appliance')['energy_consumption_kWh'].mean()

plt.figure(figsize=(10, 6))
appliance_energy.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Consumo de EnergÃ­a por Tipo de ElectrodomÃ©stico')
plt.xlabel('Tipo de ElectrodomÃ©stico')
plt.ylabel('Consumo de EnergÃ­a (kWh)')
plt.xticks(rotation=45)
plt.show()
```
