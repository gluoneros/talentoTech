```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('dataset_renovables_sintetico.csv')
data
```

```python
data.describe()
```

```python
total_de_celdas = len(data.columns)*len(data)
celdas_nulas = data.isnull().sum().sum()
```

```python
porcentaje = (celdas_nulas/total_de_celdas)*100
porcentaje
```

```python
data.shape[0]*data.shape[1]
```

```python
data['Producción (MW)'].fillna(data['Producción (MW)'].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Temperatura (°C)'].fillna(data['Temperatura (°C)'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
tabla_nueva = data.dropna(subset=['Temperatura (°C)','Región','Luz Solar (Horas)'])
```

```python
X = pd.get_dummies(tabla_nueva[['Temperatura (°C)', 'Región']])
y = tabla_nueva['Luz Solar (Horas)']
```

```python
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
tabla_nulos_luz_solar = data[data['Luz Solar (Horas)'].isnull()]
if not tabla_nulos_luz_solar.empty:
    X_nulos_solares = pd.get_dummies(tabla_nulos_luz_solar[['Temperatura (°C)', 'Región']])
    predicciones = modelo.predict(X_nulos_solares)
    data.loc[data['Luz Solar (Horas)'].isnull(),'Luz Solar (Horas)'] = predicciones
```
