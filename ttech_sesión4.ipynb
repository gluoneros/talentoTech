```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
temp = np.arange(-10,45,0.01)
temp
```

```python
plt.plot(temp,norm.pdf(temp))
plt.show()
```

```python
ejes = plt.axes()
plt.plot(temp,norm.pdf(temp))
plt.plot(temp,norm.pdf(temp, 10, 4))
ejes.grid()
ejes.set_xticks([-10,0,10,20])
ejes.set_yticks([0,0.05,0.1,0.15,0.2])
ejes.set_xlabel('Temperatura')
ejes.set_ylabel('Probabilidad')

plt.legend(['Temp Colombia','Temp NY'])

plt.show()
```

```python
import pandas as pd

data = pd.read_csv('datos_simulados.csv')

data
```

```python
nro_proy_en_marcha = data.groupby('Fuente de energa')['Nmero de proyectos en marcha'].sum().reset_index()
nro_proy_en_marcha
```

```python
co2_reduction_by_mineral = data.groupby('Mineral estratgico')['Reduccin de CO2 (toneladas)'].sum().reset_index()
co2_reduction_by_mineral
```

```python
data['Mineral estratgico'].nunique()
```

```python
num_categories = data['Mineral estratgico'].nunique()
colors = plt.cm.Set3(range(num_categories))
```

```python

plt.bar(co2_reduction_by_mineral['Mineral estratgico'], co2_reduction_by_mineral['Reduccin de CO2 (toneladas)'], color=colors)
plt.show()
```
