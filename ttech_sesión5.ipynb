	2.	Transformación de Datos: Combina los dos DataFrames utilizando el tiempo como clave para crear un nuevo DataFrame que contenga la producción de energía y las condiciones climáticas correspondientes.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
produccion_energia = {
    'hora': pd.date_range(start='2023-01-01', periods=10, freq='H'),
    'produccion_solar': [50, 55, 60, 80, 90, 85, 60, 40, 20, 0],
    'produccion_eolica': [30, 35, 40, 60, 70, 65, 50, 45, 30, 20]
}

condiciones_climaticas = {
    'hora': pd.date_range(start='2023-01-01', periods=10, freq='H'),
    'luz_solar': [300, 400, 500, 700, 800, 750, 500, 300, 150, 0],
    'velocidad_viento': [5, 7, 8, 10, 12, 11, 9, 6, 5, 4]
}
```

```python
df_energia = pd.DataFrame(produccion_energia)
df_clima = pd.DataFrame(condiciones_climaticas)
print(df_energia)
print(df_clima)
```

```python
df_final = pd.merge(df_energia, df_clima, on='hora')
df_final
```

```python
plt.figure(figsize=(10, 6))

#Primer gráfico
plt.subplot(2, 1, 1)
plt.plot(df_final['hora'], df_final['produccion_solar'], label='Producción Solar')
plt.plot(df_final['hora'], df_final['luz_solar'], label='Luz Solar')
plt.title('Producción Solar vs. Luz Solar')
plt.legend()

#Segundo gráfico
plt.subplot(2, 1, 2)
plt.plot(df_final['hora'], df_final['produccion_eolica'], label='Producción Eólica')
plt.plot(df_final['hora'], df_final['velocidad_viento'], label='Velocidad del Viento')
plt.title('Producción Eólica vs. Velocidad del Viento')
plt.legend()


plt.tight_layout()
plt.show()
```

# Comsumo de datos provenientes de API usando requests

```python
import json
import requests
import pandas as pd
```

```python
archivo = 'token.json'
with open(archivo, 'r') as file:
    token = json.load(file)
#token
```

```python
url = "https://api.electricitymap.org/v3/carbon-intensity/history"
```

```python
parametros = {
    'zone':'CO',
    'datetime':'2019-05-21T00:00:00Z'
}
```

```python
encabezado = {
    'auth-token': token['key']
}
```

```python
respuesta = requests.get(url, params=parametros, headers=encabezado)
```

```python
respuesta
```

```python
if respuesta.status_code == 200:
    data = respuesta.json()
    print(json.dumps(data, indent=4))
else:
    print(f"Error en la solicitud. Código de estado: {respuesta.status_code}")
```

```python
datos_historicos = pd.DataFrame(data['history'])
datos_historicos
```
